```python
import numpy as np
import pandas as pd
from bertopic import BERTopic
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from difflib import SequenceMatcher

def calculate_text_similarity(text1, text2):
    """
    Calculate similarity ratio between two texts using SequenceMatcher
    
    Parameters:
    -----------
    text1 : str
        First text
    text2 : str
        Second text
        
    Returns:
    --------
    float
        Similarity ratio between 0 and 1
    """
    if pd.isna(text1) or pd.isna(text2):
        return 0.0
    return SequenceMatcher(None, str(text1), str(text2)).ratio()

def find_representative_document_different(df, topic_model, documents, custom_words=None, n_docs=3, max_similarity=0.3):
    """
    Find the most representative documents for each topic using BERTopic.
    
    Parameters:
    -----------
    topic_model: the model
    documents : list
        List of documents (strings)
    custom_words : list, optional (default=None)
        List of custom words to add to the vocabulary, such as ["stance_anti", "stance_pro"]
    n_docs : int, optional (default=3)
        Number of representative documents to return per topic
    max_similarity : float, optional (default=0.3)
        Maximum allowed similarity between any pair of representative documents
        
    Returns:
    --------
    dict
        Dictionary mapping topic IDs to their most representative documents (list)
    pandas.DataFrame
        DataFrame with topic modeling results including documents, topics, and probabilities
    """
    
    print("Finding representative documents for each topic...")
    # For each topic (excluding outliers with topic ID -1)
    for topic_id in df['Topic'].unique():
        if topic_id == -1:  # Skip outlier topic
            continue
            
        # Get the documents belonging to this topic
        topic_docs = df[df['Topic'] == topic_id]['Document'].tolist()
        
        if not topic_docs:
            continue
            
        # Get the topic words
        topic_words = [word for word, _ in topic_model.get_topic(topic_id)]
        topic_words_str = " ".join(topic_words)
            
        # If we have fewer documents than requested, return all of them
        if len(topic_docs) <= n_docs:
            representative_docs[topic_id] = topic_docs
            continue
            
        # Get embeddings for documents and topic words
        sentence_model = SentenceTransformer(model_name)
        doc_embeddings = sentence_model.encode(topic_docs)
        topic_embedding = sentence_model.encode([topic_words_str])[0]
        
        # Calculate similarity between each document and the topic
        similarities = cosine_similarity([topic_embedding], doc_embeddings)[0]
        
        # Create a list of (document, similarity) tuples and sort by similarity (descending)
        doc_sim_pairs = [(doc, sim) for doc, sim in zip(topic_docs, similarities)]
        doc_sim_pairs.sort(key=lambda x: x[1], reverse=True)
        
        # Initialize the list of representative documents with the most representative one
        selected_docs = [doc_sim_pairs[0][0]]
        doc_candidates = doc_sim_pairs[1:]
        
        # Try to find additional documents with similarity < max_similarity
        while len(selected_docs) < n_docs and doc_candidates:
            next_candidate, next_sim = doc_candidates.pop(0)
            
            # Check if this candidate is dissimilar enough from all selected docs
            is_dissimilar = True
            for selected_doc in selected_docs:
                text_sim = calculate_text_similarity(next_candidate, selected_doc)
                if text_sim >= max_similarity:
                    is_dissimilar = False
                    break
            
            if is_dissimilar:
                selected_docs.append(next_candidate)
        
        representative_docs[topic_id] = selected_docs
    
    print(f"Found representative documents for {len(representative_docs)} topics")
    return representative_docs
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

def find_representative_document(documents, nr_topics, topic_model,topics,probs):
    """
    Find the most representative document for each topic using BERTopic.
    
    Parameters:
    -----------
    documents : list
        List of documents (strings)
    nr_topics : int, optional (default=10)
        Number of topics to extract
    model_name : str, optional (default="all-MiniLM-L6-v2")
        Name of the sentence-transformer model to use
        
    Returns:
    --------
    dict
        Dictionary mapping topic IDs to their most representative document
    pandas.DataFrame
        DataFrame with topic modeling results including documents, topics, and probabilities
    """
   
    # Get topic info
    topic_info = topic_model.get_topic_info()
    
    # Create a dataframe with documents and their topics
    df = pd.DataFrame({
        'Document': documents,
        'Topic': topics,
        'Probability':  [max(prob) if isinstance(prob, np.ndarray) else prob for prob in probs]
    })
    
    # Initialize dictionary to store representative documents
    representative_docs = {}
    
    print("Finding representative documents for each topic...")
    # For each topic (excluding outliers with topic ID -1)
    for topic_id in df['Topic'].unique():
        if topic_id == -1:  # Skip outlier topic
            continue
            
        # Get the documents belonging to this topic
        topic_docs = df[df['Topic'] == topic_id]['Document'].tolist()
        
        if not topic_docs:
            continue
            
        # Get the topic words
        topic_words = [word for word, _ in topic_model.get_topic(topic_id)]
        topic_words_str = " ".join(topic_words)
        
        # If we have a single document, it's the representative one
        if len(topic_docs) == 1:
            representative_docs[topic_id] = topic_docs[0]
            continue
            
        # Get embeddings for documents and topic words
        sentence_model = SentenceTransformer(model_name)
        doc_embeddings = sentence_model.encode(topic_docs)
        topic_embedding = sentence_model.encode([topic_words_str])[0]
        
        # Calculate similarity between each document and the topic
        similarities = cosine_similarity([topic_embedding], doc_embeddings)[0]
        
        # Find the document with the highest similarity
        most_representative_idx = np.argmax(similarities)
        representative_docs[topic_id] = topic_docs[most_representative_idx]
    
    print(f"Found representative documents for {len(representative_docs)} topics")
    return representative_docs, df

def show_topic_summary(topic_model, representative_docs):
    """
    Print a summary of each topic with its words and representative document.
    
    Parameters:
    -----------
    topic_model : BERTopic
        Fitted BERTopic model
    representative_docs : dict
        Dictionary mapping topic IDs to their most representative document
    """
    topic_info = topic_model.get_topic_info()
    
    # For each topic (excluding outliers)
    for topic_id, topic_name in zip(topic_info['Topic'], topic_info['Name']):
        if topic_id == -1:  # Skip outlier topic
            continue
            
        print(f"\nTopic {topic_id}: {topic_name}")
        print("="*50)
        
        # Print top words
        topic_words = topic_model.get_topic(topic_id)
        print("Top words:")
        for word, weight in topic_words[:10]:
            print(f"  - {word} ({weight:.4f})")
        
        # Print representative document
        if topic_id in representative_docs:
            print("\nRepresentative document:")
            doc = representative_docs[topic_id]
            # Truncate long documents for display
            if len(doc) > 300:
                print(doc[:300] + "...")
            else:
                print(doc)
        else:
            print("\nNo representative document found.")
        print("-"*50)
```

```python
def get_topic_keywords_long_format(topic_model, n_words=10, include_weights=True):
    """
    Get topic keywords in a long format with multiple rows per topic (one for each keyword).
    
    Parameters:
    -----------
    topic_model : BERTopic
        A fitted BERTopic model
    n_words : int, optional (default=10)
        Number of keywords to extract per topic
    include_weights : bool, optional (default=True)
        Whether to include keyword weights
    
    Returns:
    --------
    pandas.DataFrame
        DataFrame with one row per topic-keyword combination
    """
    # Get the wide format DataFrame
    wide_df = get_topic_keywords_df(topic_model, n_words, include_weights)
    
    # Select the base columns (always include these)
    base_columns = ['Topic', 'Name']
    if 'Count' in wide_df.columns:
        base_columns.extend(['Count'])
    
    # Get all keyword columns
    keyword_columns = [col for col in wide_df.columns if col.startswith('Keyword_')]
    
    # Get all weight columns if included
    weight_columns = [col for col in wide_df.columns if col.startswith('Weight_')] if include_weights else []
    
    # List to store unpivoted data
    long_data = []
    
    # For each row in the wide DataFrame
    for _, row in wide_df.iterrows():
        # Get the base data for this topic
        base_data = {col: row[col] for col in base_columns}
        
        # For each keyword
        for i, keyword_col in enumerate(keyword_columns):
            # Copy the base data for this keyword
            keyword_data = base_data.copy()
            
            # Add keyword information
            keyword_data['Keyword'] = row[keyword_col]
            keyword_data['Rank'] = i + 1
            
            # Add weight information if included
            if include_weights and i < len(weight_columns):
                keyword_data['Weight'] = row[weight_columns[i]]
            
            long_data.append(keyword_data)
    
    # Create the long format DataFrame
    long_df = pd.DataFrame(long_data)
    
    return long_df
```

```python
def get_topic_keywords_df(topic_model, n_words=10, include_weights=True, include_topic_info=True):
    """
    Extract topic keywords from a BERTopic model and return them as a DataFrame.
    
    Parameters:
    -----------
    topic_model : BERTopic
        A fitted BERTopic model
    n_words : int, optional (default=10)
        Number of keywords to extract per topic
    include_weights : bool, optional (default=True)
        Whether to include keyword weights in the DataFrame
    include_topic_info : bool, optional (default=True)
        Whether to include additional topic information (size, count)
    
    Returns:
    --------
    pandas.DataFrame
        DataFrame containing topic keywords and optionally weights and additional info
    """
    # Get the topic info DataFrame (contains topic sizes, etc.)
    topic_info = topic_model.get_topic_info()
    
    # Create a list to store data for each topic
    topic_keywords_data = []
    
    # Process each topic (excluding outlier topic -1)
    for topic_id in sorted(topic_info[topic_info['Topic'] != -1]['Topic']):
        # Get topic words with their weights
        topic_words_weights = topic_model.get_topic(topic_id)[:n_words]
        
        # Separate words and weights
        words = [word for word, _ in topic_words_weights]
        weights = [weight for _, weight in topic_words_weights]
        
        # Basic topic data (always included)
        topic_data = {'Topic': topic_id}
        
        # Add topic name if available in topic_info
        topic_name = topic_info[topic_info['Topic'] == topic_id]['Name'].values[0]
        topic_data['Name'] = topic_name
        
        # Add additional topic info if requested
        if include_topic_info:
            topic_data['Count'] = topic_info[topic_info['Topic'] == topic_id]['Count'].values[0]
            #topic_data['Size (%)'] = topic_info[topic_info['Topic'] == topic_id]['Representation'].values[0]
        
        # Add keywords
        for i, word in enumerate(words):
            topic_data[f'Keyword_{i+1}'] = word
            
            # Add weights if requested
            if include_weights:
                topic_data[f'Weight_{i+1}'] = weights[i]
        
        topic_keywords_data.append(topic_data)
    
    # Create DataFrame from the collected data
    topic_keywords_df = pd.DataFrame(topic_keywords_data)
    
    return topic_keywords_df
```
